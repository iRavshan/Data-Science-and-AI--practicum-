```python
import pandas as pd
import numpy as np
```

```python
tashkent = pd.read_html('https://en.tutiempo.net/climate/ws-384570.html', index_col=0)
df = tashkent[3][tashkent[3].index > 1990]
df.head()
```

```python
df = df.replace('-', np.nan)
df.head(10)
```

```python
df.isnull().sum() / len(df) * 100 
```

```python
df.drop('PP', axis=1, inplace=True)
df.head()
```

```python
df.fillna(method='ffill', inplace=True)
df.head(10)
```

```python
df.to_csv('tashkent_weather.csv')
```
